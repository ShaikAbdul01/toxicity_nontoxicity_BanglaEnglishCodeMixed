```python
# Cell 1: First, let's explore your dataset structure properly
import pandas as pd
import numpy as np
from sklearn.metrics import cohen_kappa_score
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("code_mixed_dataset.csv")

print("=== DATASET STRUCTURE ANALYSIS ===")
print(f"Dataset shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
print("\nFirst few rows:")
print(df.head(10))

# Check for multiple annotations per sentence
print(f"\n=== MULTIPLE ANNOTATIONS ANALYSIS ===")
print(f"Total rows: {len(df)}")
print(f"Unique sentences: {df['sentence'].nunique()}")

# Count annotations per sentence
sentence_counts = df['sentence'].value_counts()
print(f"\nAnnotations per sentence distribution:")
print(sentence_counts.value_counts().sort_index())

# Show sentences with multiple annotations
multi_annotated = sentence_counts[sentence_counts > 1]
print(f"\nSentences with multiple annotations: {len(multi_annotated)}")
if len(multi_annotated) > 0:
    print("Examples of multi-annotated sentences:")
    for sent in multi_annotated.index[:3]:
        annotations = df[df['sentence'] == sent]['label(0->Tocxic & 1 -> Non-toxic)'].values
        print(f"Sentence: {sent[:80]}...")
        print(f"Annotations: {annotations}")
```

```python
# Cell 2: Check if you have a proper multi-annotator dataset structure
# Based on the error, it seems you might have a different structure

# Option 1: If each row is one annotation and sentences are repeated
if df['sentence'].nunique() < len(df):
    print("Dataset appears to have multiple annotations per sentence (repeated sentences)")
    
    # Let's see the actual structure
    sample_sentences = df['sentence'].value_counts().head(5)
    for sentence, count in sample_sentences.items():
        print(f"\nSentence (first 100 chars): {sentence[:100]}...")
        print(f"Number of annotations: {count}")
        annotations = df[df['sentence'] == sentence]['label(0->Tocxic & 1 -> Non-toxic)'].values
        print(f"Labels: {annotations}")
        
else:
    print("Dataset appears to have single annotation per sentence")
    print("This dataset might not be structured for multi-annotator analysis")
```

```python
# Cell 3: Alternative approach - check if you have separate columns for each annotator
print("Checking for annotator columns in dataset...")

# Look for columns that might contain annotator data
annotator_columns = [col for col in df.columns if 'annotator' in col.lower() or 'label' in col.lower()]
print(f"Potential label/annotator columns: {annotator_columns}")

if len(annotator_columns) >= 3:
    print("Found multiple annotator columns!")
    for col in annotator_columns:
        print(f"{col}: {df[col].value_counts().to_dict()}")
else:
    print("No separate annotator columns found. Dataset might have single annotation per row.")
```

```python
# Cell 4: Let's handle both scenarios - restructure based on actual data format

def analyze_annotation_quality(df):
    """Analyze annotation quality based on actual dataset structure"""
    
    # Clean the data first
    df_clean = df.dropna().copy()
    df_clean.rename(columns={'label(0->Tocxic & 1 -> Non-toxic)': 'label'}, inplace=True)
    df_clean['label'] = pd.to_numeric(df_clean['label'], errors='coerce')
    df_clean = df_clean.dropna(subset=['label']).copy()
    df_clean['label'] = df_clean['label'].astype(int)
    
    print(f"Clean dataset size: {len(df_clean)}")
    
    # SCENARIO 1: Multiple rows per sentence (multiple annotators)
    sentence_counts = df_clean['sentence'].value_counts()
    max_annotators = sentence_counts.max()
    
    if max_annotators >= 2:
        print(f"\n=== MULTI-ANNOTATOR ANALYSIS (Max {max_annotators} annotators per sentence) ===")
        return analyze_multi_annotator_structure(df_clean, max_annotators)
    
    # SCENARIO 2: Single annotation per sentence
    else:
        print("\n=== SINGLE ANNOTATOR DATASET ===")
        print("This dataset has only one annotation per sentence.")
        print("To perform multi-annotator analysis, you need multiple annotations per sentence.")
        return analyze_single_annotator_quality(df_clean)

def analyze_multi_annotator_structure(df_clean, max_annotators):
    """Analyze when we have multiple annotations per sentence"""
    
    # Create annotator IDs
    df_clean['annotator_id'] = df_clean.groupby('sentence').cumcount() + 1
    
    # Pivot to wide format
    pivot_df = df_clean.pivot_table(
        index='sentence', 
        columns='annotator_id', 
        values='label', 
        aggfunc='first'
    ).reset_index()
    
    # Rename columns based on actual number of annotators
    available_annotators = list(range(1, pivot_df.shape[1]))
    pivot_df.columns = ['sentence'] + [f'annotator_{i}' for i in available_annotators]
    
    print(f"Available annotators: {available_annotators}")
    print(f"Shape of pivot table: {pivot_df.shape}")
    
    # Only use sentences with at least 2 annotators for agreement analysis
    min_annotators = 2
    complete_annotations = pivot_df.dropna(thresh=min_annotators+1)  # +1 for sentence column
    
    print(f"Sentences with at least {min_annotators} annotators: {len(complete_annotations)}")
    
    if len(complete_annotations) < 2:
        print("Not enough sentences with multiple annotations for agreement analysis")
        return None
    
    return complete_annotations, available_annotators

def analyze_single_annotator_quality(df_clean):
    """Provide basic quality analysis for single-annotator dataset"""
    print("Basic dataset quality analysis:")
    print(f"Total sentences: {len(df_clean)}")
    print(f"Toxic (0): {(df_clean['label'] == 0).sum()} ({(df_clean['label'] == 0).mean()*100:.1f}%)")
    print(f"Non-toxic (1): {(df_clean['label'] == 1).sum()} ({(df_clean['label'] == 1).mean()*100:.1f}%)")
    
    # Check for potential data quality issues
    duplicate_sentences = df_clean[df_clean.duplicated('sentence', keep=False)]
    if len(duplicate_sentences) > 0:
        print(f"\nWARNING: Found {len(duplicate_sentences)} duplicate sentences with potentially different labels")
    
    return df_clean

# Run the analysis
result = analyze_annotation_quality(df)
```

```python
# Cell 5: Calculate agreement metrics based on actual annotator count
if result is not None:
    if isinstance(result, tuple):
        complete_annotations, available_annotators = result
        print(f"\n=== CALCULATING AGREEMENT METRICS ===")
        print(f"Using {len(available_annotators)} annotators: {available_annotators}")
        
        # Calculate pairwise agreement for available annotators
        if len(available_annotators) >= 2:
            pairwise_kappas = {}
            for i in range(len(available_annotators)):
                for j in range(i+1, len(available_annotators)):
                    ann1 = f'annotator_{available_annotators[i]}'
                    ann2 = f'annotator_{available_annotators[j]}'
                    
                    # Only calculate if both columns exist and have data
                    if ann1 in complete_annotations.columns and ann2 in complete_annotations.columns:
                        valid_data = complete_annotations[[ann1, ann2]].dropna()
                        if len(valid_data) >= 2:
                            kappa = cohen_kappa_score(valid_data[ann1], valid_data[ann2])
                            pairwise_kappas[f"{ann1}_vs_{ann2}"] = kappa
                            print(f"Cohen's Kappa ({ann1} vs {ann2}): {kappa:.4f}")
            
            if pairwise_kappas:
                avg_kappa = np.mean(list(pairwise_kappas.values()))
                print(f"\nAverage Cohen's Kappa: {avg_kappa:.4f}")
                
                # Interpretation
                if avg_kappa < 0:
                    interpretation = "Poor agreement"
                elif avg_kappa <= 0.20:
                    interpretation = "Slight agreement"
                elif avg_kappa <= 0.40:
                    interpretation = "Fair agreement"
                elif avg_kappa <= 0.60:
                    interpretation = "Moderate agreement"
                elif avg_kappa <= 0.80:
                    interpretation = "Substantial agreement"
                else:
                    interpretation = "Almost perfect agreement"
                
                print(f"Interpretation: {interpretation}")
        
        # Analyze agreement levels
        if len(available_annotators) >= 2:
            print(f"\n=== AGREEMENT LEVEL ANALYSIS ===")
            
            # For each sentence, count how many annotators agree
            annotation_cols = [f'annotator_{i}' for i in available_annotators]
            
            complete_annotations['agreement_count'] = complete_annotations[annotation_cols].apply(
                lambda row: row.nunique() if not row.isna().any() else np.nan, 
                axis=1
            )
            
            complete_annotations = complete_annotations.dropna(subset=['agreement_count'])
            
            agreement_stats = complete_annotations['agreement_count'].value_counts().sort_index()
            
            print("Agreement Distribution:")
            for count, freq in agreement_stats.items():
                unique_labels = len(available_annotators) + 1 - count
                percentage = (freq / len(complete_annotations)) * 100
                print(f"  {unique_labels} unique labels: {freq} sentences ({percentage:.1f}%)")
    
    else:
        print("\nSingle annotator dataset - cannot calculate inter-annotator agreement")
else:
    print("No valid multi-annotator data found for analysis")
```

```python
# Cell 6: Create visualizations based on actual data structure
plt.figure(figsize=(15, 5))

if result is not None and isinstance(result, tuple):
    complete_annotations, available_annotators = result
    
    # Plot 1: Label distribution by annotator
    plt.subplot(1, 3, 1)
    annotator_data = []
    for ann in available_annotators:
        col_name = f'annotator_{ann}'
        if col_name in complete_annotations.columns:
            toxic_pct = (complete_annotations[col_name] == 0).mean() * 100
            non_toxic_pct = (complete_annotations[col_name] == 1).mean() * 100
            annotator_data.append([f'Ann{ann}', toxic_pct, non_toxic_pct])
    
    if annotator_data:
        annotator_df = pd.DataFrame(annotator_data, columns=['Annotator', 'Toxic%', 'Non-toxic%'])
        annotator_df.set_index('Annotator')[['Toxic%', 'Non-toxic%']].plot(kind='bar', stacked=True, 
                                                                          color=['lightcoral', 'lightgreen'])
        plt.title('Label Distribution by Annotator')
        plt.ylabel('Percentage')
        plt.xticks(rotation=45)
    
    # Plot 2: Agreement distribution
    plt.subplot(1, 3, 2)
    if 'agreement_count' in complete_annotations.columns:
        agreement_stats = complete_annotations['agreement_count'].value_counts().sort_index()
        labels = [f'{len(available_annotators)+1-count} labels' for count in agreement_stats.index]
        plt.pie(agreement_stats.values, labels=labels, autopct='%1.1f%%')
        plt.title('Agreement Level Distribution')
    
    # Plot 3: Majority vote distribution
    plt.subplot(1, 3, 3)
    if len(available_annotators) >= 2:
        # Calculate majority vote
        annotation_cols = [f'annotator_{i}' for i in available_annotators]
        complete_annotations['majority_vote'] = complete_annotations[annotation_cols].mode(axis=1)[0]
        
        majority_counts = complete_annotations['majority_vote'].value_counts()
        plt.pie(majority_counts.values, labels=['Toxic', 'Non-toxic'], autopct='%1.1f%%', 
                colors=['lightcoral', 'lightgreen'])
        plt.title('Majority Vote Distribution')

plt.tight_layout()
plt.show()
```

```python
# Cell 7: Create final consolidated dataset with quality metrics
if result is not None and isinstance(result, tuple):
    complete_annotations, available_annotators = result
    
    print("=== CREATING FINAL DATASET ===")
    
    # Calculate majority vote
    annotation_cols = [f'annotator_{i}' for i in available_annotators]
    complete_annotations['majority_vote'] = complete_annotations[annotation_cols].mode(axis=1)[0]
    
    # Calculate agreement level
    complete_annotations['agreement_level'] = complete_annotations[annotation_cols].apply(
        lambda row: row.nunique() if not row.isna().any() else np.nan, axis=1
    )
    
    complete_annotations = complete_annotations.dropna(subset=['agreement_level'])
    
    # Create final dataset
    final_dataset = complete_annotations[['sentence', 'majority_vote', 'agreement_level']].copy()
    final_dataset.rename(columns={'majority_vote': 'consensus_label'}, inplace=True)
    
    # Add individual annotations
    for ann in available_annotators:
        col_name = f'annotator_{ann}'
        final_dataset[col_name] = complete_annotations[col_name]
    
    # Add confidence score (higher = more agreement)
    final_dataset['confidence_score'] = (len(available_annotators) - final_dataset['agreement_level'] + 1) / len(available_annotators)
    
    print(f"Final dataset created with {len(final_dataset)} sentences")
    print(f"Consensus label distribution:")
    print(final_dataset['consensus_label'].value_counts().sort_index())
    
    # Save final dataset
    final_dataset.to_csv('consolidated_dataset_with_quality.csv', index=False)
    print("\nFinal dataset saved as 'consolidated_dataset_with_quality.csv'")
    
    # Show sample
    print("\nSample of final dataset:")
    print(final_dataset.head(10))
    
else:
    print("Cannot create consolidated dataset - insufficient multi-annotator data")
```

```python
# Cell 8: Summary report
print("=== ANNOTATION QUALITY SUMMARY ===")

if result is not None:
    if isinstance(result, tuple):
        complete_annotations, available_annotators = result
        print(f"Number of annotators: {len(available_annotators)}")
        print(f"Sentences with complete annotations: {len(complete_annotations)}")
        
        if len(available_annotators) >= 2:
            # Calculate average agreement
            annotation_cols = [f'annotator_{i}' for i in available_annotators]
            unanimous_count = complete_annotations[annotation_cols].apply(
                lambda row: row.nunique() == 1 if not row.isna().any() else False, axis=1
            ).sum()
            
            print(f"Unanimous agreements: {unanimous_count} ({(unanimous_count/len(complete_annotations))*100:.1f}%)")
            
    else:
        print("Single annotator dataset")
        print(f"Total sentences: {len(result)}")
        print(f"Toxic: {(result['label'] == 0).sum()} ({(result['label'] == 0).mean()*100:.1f}%)")
        print(f"Non-toxic: {(result['label'] == 1).sum()} ({(result['label'] == 1).mean()*100:.1f}%)")

print("\n=== RECOMMENDATIONS ===")
if result is not None and isinstance(result, tuple) and len(available_annotators) < 3:
    print("⚠️  Limited multi-annotator data detected")
    print("• Consider collecting more annotations per sentence")
    print("• Aim for at least 3 annotators per sentence for reliable IAA")
    print("• Current analysis uses available data but results may be limited")
else:
    print("✓ Multi-annotator analysis completed successfully")
```
